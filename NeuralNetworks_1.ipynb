```python
import numpy as np
```

```python
X = np.array([[1,2,3,2,0,1],[4,5,6,1,1,0],[1,2,3,2,5,3],[2,1,3,4,5,6],[2,3,4,2,1,6],[2,7,3,4,5,2]])
Y = np.array([1,0,0,1,0,1])

```

```python
import time
from sklearn.metrics import accuracy_score

class Dense_Tanh_1hidden():
  def __init__(self):
    return
  def fit(self,X,Y,neurons=4,epochs=10000,learning_rate=0.01):
    m = X.shape[0]
    self.W_1 = np.random.randn(neurons,X.shape[1]) * 0.01
    self.b_1 = np.zeros((neurons,1))
    self.W_2 = np.random.randn(1,neurons)*0.01
    self.b_2 = np.zeros((1,1))
    for x in range(epochs):
      Z_1 = np.dot(self.W_1,X.T)+self.b_1
      A_1 = (np.exp(Z_1) - np.exp(-Z_1))/(np.exp(Z_1)+np.exp(-Z_1))
      Z_2 = np.dot(self.W_2,A_1) + self.b_2
      A_2 = 1/(1+np.exp(-Z_2))
      loss = np.sum(-(Y*np.log(A_2) + (1-Y)*np.log(1-A_2)))
      a = (A_2>=0.5).astype(int).reshape(X.shape[1],)
      print(f"loss: {loss}, accuracy: {accuracy_score(Y,a)*100}%")
      dZ_2 = A_2-Y
      dW_2 = 1/m*np.dot(dZ_2,A_1.T)
      db_2 = 1/m*np.sum(dZ_2,axis=1,keepdims=True)
      dgz = 1-np.power(A_1,2)
      dZ_1 = np.dot(self.W_2.T,dZ_2)*dgz
      dW_1 = 1/m*np.dot(dZ_1,X)
      db_1 = 1/m*np.sum(dZ_1,axis=1,keepdims=True)
      self.W_1 = self.W_1 - learning_rate*dW_1
      self.b_1 = self.b_1 - learning_rate*db_1
      self.W_2 = self.W_2 - learning_rate*dW_2
      self.b_2 = self.b_2 - learning_rate*db_2
  def predict(self,X,treshold=0.5):
      Z_1 = np.dot(self.W_1,X.T)+self.b_1
      A_1 = (np.exp(Z_1) - np.exp(-Z_1))/(np.exp(Z_1)+np.exp(-Z_1))
      Z_2 = np.dot(self.W_2,A_1) + self.b_2
      A_2 = 1/(1+np.exp(-Z_2))
      print((A_2>=treshold).astype(int))



```

```python
model = Dense_Tanh_1hidden()
```

```python
model.fit(X,Y,neurons=1000)
```

```python
model.predict(np.array([[4,5,6,1,1,0]]),treshold=0.5)
```
